OneHotEnconder

```python
data_master_ohe=pd.get_dummies(data_frame_knn)
```

```python
data_master_ohe['Resulta']=data_master_ohe['Resultado']
```

```python
data_master_ohe.drop(columns='Resultado',axis=1,
                     inplace=True)
```

```python
model_machine_knn=KNeighborsClassifier(n_neighbors=3,metric='manhattan',n_jobs=1)
```

```python
scaler = StandardScaler()
data_master_ohe_scale = pd.DataFrame(scaler.fit_transform(data_master_ohe), columns=data_master_ohe.columns)
```

```python
data_master_ohe_scale['Resulta']=data_master_ohe_scale['Resulta'].astype(np.int64)
```

```python
feature_response=data_master_ohe_scale.columns.tolist()
```

```python
feature_response
```

```python
X_train,X_test,y_train,y_test=train_test_split(
    data_master_ohe_scale[feature_response[:-1]].values,
    data_master_ohe_scale['Resulta'].values,
    test_size=0.2,
    random_state=24
)
```

```python
model_machine_knn.fit(X_train,y_train)
```

```python
y_pred_score=model_machine_knn.score(
    X_test,y_test
)
```

```python
distances, indices = model_machine_knn.kneighbors(X_test[:1], n_neighbors=3)

# Mostrar os índices e distâncias dos vizinhos mais próximos
print(f"Índices dos vizinhos mais próximos: {indices}")
print(f"Distâncias dos vizinhos mais próximos: {distances}")
```

```python
distances, indices = model_machine_knn.kneighbors(X_test, n_neighbors=3)
neighbors_labels = y_train[indices]
majority_classes = np.apply_along_axis(lambda x: np.bincount(x).argmax(), axis=1, arr=neighbors_labels)

# Mostrando as classes majoritárias
print(f"Classes majoritárias: {majority_classes}")
```
